# TFIDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documentA = 'the man went out for a walk'
documentB = 'the children sat around the fire'
```

```python
def vectorizer():
    
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([documentA, documentB])
    feature_names = vectorizer.get_feature_names()
    vec = vectors.toarray()    
```

## Building Machine Learning Model

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.datasets import fetch_20newsgroups
```

```python
# All data
newsgroups_train = fetch_20newsgroups(subset='train',
                                      remove=('headers', 'footers', 'quotes'))
newsgroups_test = fetch_20newsgroups(subset='test',
                                     remove=('headers', 'footers', 'quotes'))
```

```python
# Create our vectorizer
vec_ = CountVectorizer() # object of the class CountVectorizer
```

```python
newsgroups_train.data  # data
```

```python
newsgroups_train.target_names # labels
```

```python
newsgroups_train.target # encoded labels # --------  y
```

```python
# Get the training vectors
vectors = vec_.transform(newsgroups_train.data) # ------ X
```

#### Decision Tree Model

```python
# Build the classifier
clf = DecisionTreeClassifier()
clf.fit(vectors, newsgroups_train.target)
```

```python
# Get the test vectors
vectors_test = vectorizer.fit_transform(newsgroups_test.data).toarray()
```

```python
# Predict and score the vectors
pred = clf.predict(vectors_test)
acc_score = metrics.accuracy_score(newsgroups_test.target, pred)
f1_score = metrics.f1_score(newsgroups_test.target, pred, average='macro')
print('Total accuracy classification score: {}'.format(acc_score))
print('Total F1 classification score: {}'.format(f1_score))
```

#### Naive Bayes Classifier

```python
# Get the training vectors
vectors = vectorizer.fit_transform(newsgroups_train.data)
# Build the classifier
clf = MultinomialNB()
#  Train the classifier
clf.fit(vectors, newsgroups_train.target)
```

```python
# Get the test vectors
vectors_test = vectorizer.transform(newsgroups_test.data)
```

```python
# Predict and score the vectors
pred = clf.predict(vectors_test)
acc_score = metrics.accuracy_score(newsgroups_test.target, pred)
f1_score = metrics.f1_score(newsgroups_test.target, pred, average='macro')

print('Total accuracy classification score: {}'.format(acc_score))
print('Total F1 classification score: {}'.format(f1_score))
```

## SPAM CLASSIFIER

```python
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/skesh/Desktop/Trees/week 5/spam.csv", encoding = "latin-1")
```

```python
data = data[['v1', 'v2']]

data = data.rename(columns = {'v1': 'label', 'v2': 'text'})
data
```

```python
def review_messages(msg):
    # converting messages to lowercase
    msg = msg.lower()
    return msg
```

```python
# Processing text messages
data['text'] = data['text'].apply(review_messages)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train test split 
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['label'], test_size = 0.1, random_state = 1)
```

```python
# training vectorizer
vectorizer = TfidfVectorizer()

#creating a vector
X_train = vectorizer.fit_transform(X_train)  # ---------------------------
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
clf = MultinomialNB(alpha = 0.1)
#  Train the classifier
clf.fit(X_train, y_train)
```

```python
#creating a vector
X_test_vec = vectorizer.transform(X_test)  # ---------------------------

y_pred = clf.predict(X_test_vec)
```
