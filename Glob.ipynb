```python
# import necessary libraries
import pandas as pd
import os
import glob
from os import listdir
from os.path import isfile, join
```

```python
def prod_market(folder_name,df_val,path,sheetname):

    df = pd.read_excel(path,sheet_name=sheetname)

    df_val["Product_Name"] = df.columns[1]
    df_val["Product_Code"] = df.loc[0,df.columns[1]]
    df_val["Market"] = df.loc[1,df.columns[1]]
    df_val["Plant"] = df.loc[1,df.columns[3]]
    
    #Creating a column for folder name
    df_val["Folder Name"] = folder_name
    
    df_val = df_val[['Folder Name','Product_Name','Product_Code', 'Market', 'Plant','CQA','USL', 'LSL', 'UCL', 'LCL', 'Average', 'SD']]
    
    
    
    return df_val


def cqa(path,sheetname):

    df = pd.ExcelFile(path)    
    df = df.parse(sheetname, header=None)
    df.columns = ["Col_"+str(i) for i in range(1,df.shape[1]+1)]

    df_val = df.head(6).set_index("Col_1").T.reset_index(drop=True)
    CQA = df.loc[7].values[1:].tolist()
    df_val["CQA"] = CQA
    
    return df_val
```

```python
def get_csv(folder_path,sub_folder_name):
    
    path = folder_path[:-1] + sub_folder_name + '/*'
    
    sub_folder_items = glob.glob(path)
    
    mother_df = pd.DataFrame()
    
    for files in sub_folder_items:
        
        filename = files.split("\\")[1]
        file_path = path[:-1]+filename
        
        df = pd.read_excel(file_path)

        mother_df = pd.concat([mother_df,df])
        
        return mother_df
```
